```python
df.drop(index=245, inplace=True)
df.reset_index(inplace=True)
```

```python
lat_list = []
long_list = []
for index, row in df.iterrows():
    try:
        
        location = geolocator.geocode(f"{row['TERMINAL CITY/TOWN/LOCALITY']}, {row['TERMINAL COUNTRY']}")
        lat = location.latitude
        long = location.longitude
        lat_list.append(lat)
        long_list.append(long)
    except AttributeError:
        print(f"Problem index: {index}. Here's more: {row['TERMINAL CITY/TOWN/LOCALITY']}, {row['TERMINAL COUNTRY']}")  
```
